## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
# print how many rows are there in the table named 'master'
QUERY = """
SELECT *
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
df["ConvertedComp"].hist()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

df["Age"].plot(kind="box",figsize=(8,12))
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

df.plot(kind="scatter",x='Age',y='WorkWeekHrs',figsize=(12,8))
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
import numpy as np
#df["Age"]=df["Age"].fillna(df["Age"].mean()).astype(int)
df.plot(kind='scatter', x="WorkWeekHrs",y='CodeRevHrs',s=df["Age"]+100,figsize=(12,8))
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY="""
    SELECT *
    FROM DatabaseDesireNextYear
    """

df1 = pd.read_sql_query(QUERY,conn)
#df1.head()
df_prop = df1.groupby('DatabaseDesireNextYear', axis=0).sum()
df_prop.sort_values("Respondent",inplace=True,ascending=False)
df_prop1=df_prop.head()
df_prop1
```

```python
# your code goes here

colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']
explode_list = [0, 0, 0, 0.1, 0.1, 0.1]
df_prop1['Respondent'].plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True       
                                 # turn off labels on pie chart
                            )
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
df["Age"] = df["Age"].fillna(df["Age"].mean()).astype(int)
#df[(df["Age"]>=30) & (df["Age"]<=35)]
#df["Age"]
df_age = df[(df["Age"]<=35) & (df["Age"]>=30)].groupby("Age",axis=0)[["WorkWeekHrs","CodeRevHrs"]].median()
df_age.head()

df_age[["WorkWeekHrs","CodeRevHrs"]].plot(kind="bar",stacked=True,figsize=(12,8))
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

df_age = df[(df["Age"]<=60) & (df["Age"]>=45)].groupby("Age",axis=0)["ConvertedComp"].median()
df_age.plot(kind='line',figsize=(12,8))
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

df_branch = df.groupby("MainBranch")["Respondent"].count()
df_branch.plot(kind="barh",figsize=(12,8))
```
